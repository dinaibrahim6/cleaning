# Data preprocessing

```python
cow_data_cat=cow_data_new[['breed_type','faecal_consistency','health_status']]
```

```python
#  we will show the number of unique values in categorical data
for feature in cow_data_cat.columns:
    unique_values = cow_data_cat[feature].unique()
    print(feature,':',cow_data_cat[feature].nunique(),unique_values)   
```

```python
# we want to see if there is an ouliers or not in our data
#the boxplot tell us that there are outliers or no
cols = cow_data_new.drop(columns=cow_data_cat)
plt.figure(figsize=(20,20))
for i,col in enumerate(cols):
    plt.subplot(4,4,i+1)
    sns.boxplot(data=cow_data_new,x=col)
    plt.xlabel(col, fontsize=15)
    plt.xticks(fontsize=10)
```

faecal_consistency by health_state

```python
plt.figure(figsize=(25,5))
ax = plt.subplot(1,2,1)
fg = sns.countplot(data=cow_data_new, hue='health_status', x='faecal_consistency')
plt.title('Faecal Consistency by Health Status')
plt.xlabel('Faecal Consistency')
plt.ylabel('Count')
plt.legend(title='Health Status')
for p in fg.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2., height+.5 , '{:.0f}'.format(height), ha="center", fontsize=8)
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=cow_data_new, x='rumen_fill', hue='health_status')
plt.title('Ruminal Fill Status by Health Status')
plt.xlabel('Ruminal Fill')
plt.ylabel('Count')

plt.legend(title='Health Status')
plt.show()
```

```python
stacked_data = cow_data_new.groupby('health_status')[['eating_duration', 'lying_down_duration']].mean()
stacked_data.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Eating and Lying Down Duration by Health Status')
plt.xlabel('Health Status')
plt.ylabel('Duration')
plt.xticks(rotation=0)
plt.legend(title='Activity', loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=cow_data_new, x='breed_type', y='milk_production', inner='quart')
plt.title('Milk Production by Breed Type')
plt.xlabel('Breed Type')
plt.ylabel('Milk Production')
plt.show()
```

this means that normal breed gives more milk than cross breed
